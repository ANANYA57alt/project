```python
def caesar_cipher(text, shift, mode='encrypt'):
    """
    Implements Caesar Cipher encryption and decryption.

    :param text: The input text to encrypt or decrypt
    :param shift: The number of positions to shift the letters
    :param mode: 'encrypt' to encrypt, 'decrypt' to decrypt
    :return: The resulting ciphertext or plaintext
    """
    result = []

    # Adjust the shift for decryption
    if mode == 'decrypt':
        shift = -shift

    for char in text:
        # Encrypt uppercase letters
        if char.isupper():
            result.append(chr((ord(char) - 65 + shift) % 26 + 65))
        # Encrypt lowercase letters
        elif char.islower():
            result.append(chr((ord(char) - 97 + shift) % 26 + 97))
        # Leave non-alphabetic characters unchanged
        else:
            result.append(char)

    return ''.join(result)


# Example Usage
plaintext = "Hello, World!"
shift = 3

# Encrypt
ciphertext = caesar_cipher(plaintext, shift, mode='encrypt')
print(f"Encrypted: {ciphertext}")  # Output: Khoor, Zruog!

# Decrypt
decrypted_text = caesar_cipher(ciphertext, shift, mode='decrypt')
print(f"Decrypted: {decrypted_text}")  # Output: Hello, World!
```

```python
def is_palindrome(s):
    # Remove spaces and convert to lowercase for case-insensitive comparison
    s = s.replace(" ", "").lower()

    # Compare the string with its reverse
    return s == s[::-1]

# Example Usage
word = "A man a plan a canal Panama"
if is_palindrome(word):
    print(f"'{word}' is a palindrome.")
else:
    print(f"'{word}' is not a palindrome.")
```

```python
def factorial(n):
    """
    Calculate the factorial of a non-negative integer n.

    :param n: Non-negative integer
    :return: Factorial of n
    """
    if n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

# Example Usage
number = 5
print(f"The factorial of {number} is {factorial(number)}.")
```

```python
def main():
    # Step 1: Get 5 items from the user and store them in a list
    items = []
    for i in range(5):
        item = input(f"Enter item {i+1}: ")
        items.append(item)

    print("\nList after adding items:")
    print(items)

    # Step 2: Allow the user to remove the first item
    if items:
        print("\nRemoving the first item...")
        removed_item = items.pop(0)
        print(f"Removed item: {removed_item}")

    # Final list
    print("\nList after removal:")
    print(items)

if __name__ == "__main__":
    main()
```

```python
def main():
    # Step 1: Get 5 integer numbers from the user and store them in a list
    numbers = []
    for i in range(5):
        while True:
            try:
                num = int(input(f"Enter integer number {i+1}: "))
                numbers.append(num)
                break
            except ValueError:
                print("Invalid input. Please enter an integer.")

    # Step 2: Display each number and its square
    print("\nNumber and their squares:")
    for number in numbers:
        print(f"Number: {number}, Square: {number ** 2}")

if __name__ == "__main__":
    main()
```

```python
def main():
    # Step 1: Create a list with some items
    items = []
    n = int(input("How many items do you want to add to the list? "))

    for i in range(n):
        item = input(f"Enter item {i+1}: ")
        items.append(item)

    print("\nOriginal List:")
    print(items)

    # Step 2: Allow the user to modify an item
    while True:
        try:
            index = int(input(f"\nEnter the index of the item you want to modify (0 to {len(items)-1}): "))
            if 0 <= index < len(items):
                new_value = input("Enter the new value: ")
                items[index] = new_value
                break
            else:
                print(f"Invalid index. Please enter a number between 0 and {len(items)-1}.")
        except ValueError:
            print("Invalid input. Please enter a valid number.")

    # Step 3: Display the modified list
    print("\nModified List:")
    print(items)

if __name__ == "__main__":
    main()
```

```python
def process_sentence():
    # Get a sentence from the user
    sentence = input("Enter a sentence: ")

    # Count the total number of characters
    total_characters = len(sentence)
    print(f"\nTotal number of characters (including spaces): {total_characters}")

    # Convert to uppercase if it's not already
    if sentence.isupper():
        print("\nThe sentence is already in uppercase.")
    else:
        sentence = sentence.upper()
        print("\nConverted sentence to uppercase:")

    print(sentence)

# Run the function
process_sentence()
```

```python
def accept_only_numbers():
    while True:
        user_input = input("Enter a number (digits only): ")
        if user_input.isdigit():
            print(f"\nYou entered a valid number: {user_input}")
            break
        else:
            print("\nInvalid input! Please enter only numeric characters (no letters or special characters).")

# Run the function
accept_only_numbers()
```

```python
def get_name_starting_with_a():
    while True:
        name = input("Enter your name (should start with 'a'): ").strip()
        if name.lower().startswith('a'):
            print(f"\nWelcome, {name}!")
            break
        else:
            print("\nError: The name must start with the letter 'a'. Please try again.")

# Run the function
get_name_starting_with_a()
```

```python
def get_name_ending_with_n():
    while True:
        name = input("Enter your name (should end with 'n'): ").strip()
        if name.lower().endswith('n'):
            print(f"\nWelcome, {name}!")
            break
        else:
            print("\nError: The name must end with the letter 'n'. Please try again.")

# Run the function
get_name_ending_with_n()
```

```python
def display_odd_index_characters():
    # Get a sentence from the user
    sentence = input("Enter a sentence: ")

    # Extract characters at odd index positions
    odd_index_chars = sentence[1::2]  # Start from index 1, take every second character

    # Display the result
    print(f"\nCharacters at odd index positions: {odd_index_chars}")

# Run the function
display_odd_index_characters()
```
