```python
#write a python program to get 5 words from the user. iterate that list to display only those word that starts with "d"
names = list()
for i in range(5):
  name = input(f"{i}. put a name:")
  names.append(name)
  for name in names:
    if name.startswith("d"):
      print(name)
```
